```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

ratingdf = pd.read_csv('/Users/rbowler/school/data301/course-project-solo_127/data/raw/rating.csv')
animedf = pd.read_csv('/Users/rbowler/school/data301/course-project-solo_127/data/raw/anime.csv')
```

```python
def load_and_process(url_or_path_to_csv_file):
    
    # Load the raw data
    animedf = pd.read_csv(url_or_path_to_csv_file)
    
    # Remove any Missing or Unknown values as well as the anime_id column
    animedf = animedf.drop(
        labels = ['anime_id'], axis = 1
        ).dropna(
        subset = ['genre', 'type', 'episodes', 'rating', 'name']
        ).loc[
        lambda row: ~row['episodes'].str.contains('Unknown')
    ]
    
    return animedf
```

```python
# Genre corresponding to Rating data
genredf = animedf[['genre', 'rating']].groupby('genre').mean().sort_values(by = ['rating'], ascending = False).reset_index()
genredf
```

```python
# Episodes corresponding to Members data
episodesdf = animedf[['episodes', 'members']].groupby('episodes').mean().sort_values(by = ['episodes'], ascending = False).reset_index()
episodesdf
```
